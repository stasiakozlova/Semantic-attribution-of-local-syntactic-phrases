```python
file = open('Текст3 для коллективного проекта 2022_tree.txt')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup (file)
```

```python
import pandas as pd
df_classes = pd.read_excel('Классы.xlsx', names =['Номер класса', 'Название класса'])
```

```python
df_classes
```

```python
print(soup.prettify())
```

```python
df_tree = pd.DataFrame(columns = ['ID', 'Лемма', 'Номер класса', 'ID хозяина', 'Морфология', 'ID кореференции'])
```

```python
df_tree
```

```python
k = 0
for se in soup.find_all('se'):
  k += 1
  id = str(k)
  for w in se.find_all('w'):
    w_id = w.get('id')
    final_id = id + '.' + w_id
    semantic_class = w.get('class')
    semantic_value = semantic_class[0]
    lemma = w.get('lemma')
    morph = w.get('morph')
    for rel in w.find_all('rel'):
      master_id = str(rel.get('id_head'))
      final_master_id = id + '.' + master_id
    final_coreference = '-'
    for coref in w.find_all('coref'):
      coreference = str(coref.get('ante'))
      final_coreference = id + '.' + coreference
    new_entry = {'ID': final_id, 'Лемма': lemma, 'Номер класса': semantic_value, 'ID хозяина': final_master_id, 'Морфология': morph, 'ID кореференции': final_coreference}
    df_tree.loc[len(df_tree)] = new_entry
```

```python
df_tree
```

```python
df_tree['Название класса'] = ""
```

```python
df_tree
```

```python
for n in range(0, len(df_tree)):
  class_value = df_tree['Номер класса'].values[n]
  row_number_list = df_classes[df_classes['Номер класса'] == class_value].index.tolist()
  if row_number_list != []:
    row_number = row_number_list[0]
    class_name = df_classes['Название класса'].values[row_number]
    df_tree.loc[n,'Название класса'] = class_name
```

```python
df_tree
```

```python
df_tree['Хозяин'] = ""
```

```python
for n in range(0, len(df_tree)):
  master_value = df_tree['ID хозяина'].values[n]
  row_number_list = df_tree[df_tree['ID'] == master_value].index.tolist()
  if row_number_list != []:
    row_number = row_number_list[0]
    master_name = df_tree['Лемма'].values[row_number]
    df_tree.loc[n,'Хозяин'] = master_name
```

```python
df_tree
```

```python
df_tree['Кореференция'] = ""
```

```python
for n in range(0, len(df_tree)):
  master_value = df_tree['ID кореференции'].values[n]
  row_number_list = df_tree[df_tree['ID'] == master_value].index.tolist()
  if row_number_list != []:
    row_number = row_number_list[0]
    master_name = df_tree['Лемма'].values[row_number]
    df_tree.loc[n,'Кореференция'] = master_name
```
